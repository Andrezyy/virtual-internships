```python
def str_to_date(x):
    return pd.Timestamp(x.business_dt)

train = pd.read_csv(r"Task3.csv")
train.business_dt = train.apply(func=str_to_date, axis=1)
#train.business_dt[0]pd.Timestamp(train.business_dt[0])
```

```python
train
```

```python
money_spent = train.groupby(['restaraunt','business_dt'],as_index=False).sum()
```

```python
plt.figure(figsize=(14,8))
c = ['blue','red','orange']
plt.title('Динамика трат денег')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
plt.xticks(rotation=45)
for i,rest in enumerate(money_spent.restaraunt.unique()):
    temp = money_spent[money_spent.restaraunt == rest]
    plt.plot_date(x=temp.business_dt, y=temp.money_spent, ls='-', color=c[i])
plt.xlabel('Date')
plt.ylabel('Money spent')
plt.legend(money_spent.restaraunt.unique())
plt.show()
```

```python
count_costumers = train.groupby(['restaraunt','business_dt'],as_index=False).count()
```

```python
plt.figure(figsize=(14,8))
c = ['blue','red','orange']
plt.title('Динамика кол-ва посетителей')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
plt.xticks(rotation=45)
for i,rest in enumerate(count_costumers.restaraunt.unique()):
    temp = count_costumers[count_costumers.restaraunt == rest]
    plt.plot_date(x=temp.business_dt, y=temp.client_id, ls='-', color=c[i])
plt.xlabel('Дата')
plt.ylabel('Кол-во посетителей')
plt.legend(money_spent.restaraunt.unique())
plt.show()
```

```python
age_dist = train.groupby(['restaraunt','age'],as_index=False).count()
```

```python
fig, axes = plt.subplots(1,3, figsize=(16,6))
fig.suptitle('Распределение возрастов по ресторанам')
plt.xticks(rotation=45)
c = ['blue','red','orange']
for i,rest in enumerate(age_dist.restaraunt.unique()):
    temp = age_dist[age_dist.restaraunt == rest]
    axes[i].set_title(rest)
    axes[i].hist(x=temp.age, weights=temp.client_id, bins=6, density=True, color=c[i])
    axes[i].set_xlabel('Возраст')
    axes[i].set_ylabel('Доля')
plt.show()
```

```python
restaraunt = count_costumers.restaraunt.unique()
c = ['blue','red','orange']
plt.figure(figsize=(14,8))
plt.title('Одновременная визуализация параметров')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
plt.xticks(rotation=45)
for i,rest in enumerate(restaraunt):
    temp_spent = money_spent[money_spent.restaraunt == rest]
    temp_count = count_costumers[count_costumers.restaraunt == rest]
    plt.plot_date(x=temp_spent.business_dt, y=temp_spent.money_spent, ls='-', color=c[i])
    plt.plot_date(x=temp_count.business_dt, y=temp_count.client_id, ls='--', color=c[i])
plt.xlabel('Date')
plt.ylabel('Custumers count')
plt.legend(['burger king [spent]','burger king [count]', 'kfc [spent]', 'kfc [count]', 'mcdonalds [spent]', 'mcdonalds [count]'])
plt.show()
```
